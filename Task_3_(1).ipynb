```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
df.head()
```

```python
df.info()
df.describe()
```

```python
df.drop(['Id', 'Species'], axis=1).hist(figsize=(10,6))
plt.suptitle("Distribution of Iris Features")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(data=df.drop(['Id'], axis=1))
plt.title("Box Plot for Iris Features")
plt.show()

```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.drop(['Id', 'Species'], axis=1).corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap  Iris Dataset")
plt.show()
```
